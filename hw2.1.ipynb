```python
import numpy as np
import scipy.io
import matplotlib.pyplot as plt

# Load the data21.mat file
mat_file_path = r'C:\Users\vanes\Desktop\portofolio\μηχανικη_μαθηση\hw2 (1)\data21.mat'
data = scipy.io.loadmat(mat_file_path)

# Print the keys to see what variables are available
print(data.keys())
```

```python
# Extract matrices A_1, A_2, B_1, B_2 from the loaded data
A1 = data['A_1']
A2 = data['A_2']
B1 = data['B_1']
B2 = data['B_2']

# Activation functions
def relu(x):
    return np.maximum(x, 0)

def sigmoid(x):
    return 1 / (1 + np.exp(-x))

# Function to generate an image from a random input Z
def generate_image():
    Z = np.random.normal(0, 1, (10, 1))  # Gaussian distribution input Z
    W1 = np.dot(A1, Z) + B1
    Z1 = relu(W1)
    W2 = np.dot(A2, Z1) + B2
    X = sigmoid(W2)
    return X

# Generate 100 images
images = [generate_image() for _ in range(100)]

# Plot the images in a 10x10 grid
fig, axes = plt.subplots(10, 10, figsize=(10, 10))
for i, ax in enumerate(axes.flat):
    img = images[i].reshape(28, 28)  # Reshape the 784x1 vector into a 28x28 image
    img = np.rot90(img)  # Rotate the image 90 degrees counterclockwise
    ax.imshow(img, cmap='gray')
    ax.axis('off')  # Hide the axes

plt.show()
```
